```python
import pandas as pd
from bigdl.chronos.data import TSDataset
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/numenta/NAB/v1.0/data/realKnownCause/nyc_taxi.csv", parse_dates=["timestamp"])
```

```python
df
```

```python
train, valid, test = TSDataset.from_pandas(df, dt_col="timestamp", target_col="value", with_split=True, val_ratio=0.1, test_ratio=0.1)
```

```python
scaler = StandardScaler()
for data in [train, valid, test]:
    data.deduplicate() \
        .impute() \
        .gen_dt_feature() \
        .scale(scaler, fit=(data is train)) \
        .roll(lookback=6, horizon=1)
```

```python
train.df
```

```python
from bigdl.chronos.forecaster.tcn_forecaster import TCNForecaster
```

```python
X, y = train.to_numpy()
```

```python
X.shape, y.shape
```

```python
tcn = TCNForecaster(past_seq_len=6, future_seq_len=1, input_feature_num=X.shape[-1], output_feature_num=y.shape[-1], seed=1)                           
```

```python
model = tcn.fit((X, y), epochs=5)
```

```python
X_test, y_test = test.to_numpy()
```

```python
prediction = tcn.predict(X_test)
```

```python
p, t = test.unscale_numpy(prediction), test.unscale_numpy(y_test)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(24,6))
plt.plot(p[:,:,0])
plt.plot(t[:,:,0])
plt.legend(["prediction (TCN)", "ground truth"])
```

```python
from bigdl.chronos.forecaster.lstm_forecaster import LSTMForecaster
```

```python
lstm = LSTMForecaster(past_seq_len=6, input_feature_num=X.shape[-1], output_feature_num=y.shape[-1], hidden_dim=32, layer_num=2, optimizer='Adam', loss='mse', lr=1e-1)
```

```python
lstm.fit((X, y), epochs=5)
```

```python
lstm.evaluate((X, y))
```

```python
prediction_lstm = lstm.predict(X_test)
```

```python
p2 = test.unscale_numpy(prediction_lstm)
```

```python
plt.figure(figsize=(24,6))
plt.plot(p2[:,:,0])
plt.plot(t[:,:,0])
plt.legend(["prediction (LSTM)", "ground truth"])
```

```python
plt.figure(figsize=(24,6))
plt.plot(p[:,:,0])
plt.plot(p2[:,:,0])
plt.plot(t[:,:,0])
plt.legend(["prediction (TCN)", "prediction (LSTM)", "ground truth"])
```
