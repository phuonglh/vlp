```python
import pandas as pd
from bigdl.chronos.data import TSDataset
from sklearn.preprocessing import StandardScaler
```

```python
featurePaths = [f"{basePath}{f}.80-19.csv" for f in ["geo", "hum", "slp", "ssm", "uvw"]]
```

```python
print(featurePaths)
```

```python
dfs = [pd.read_csv(path) for path in featurePaths]
```

```python
df = pd.concat(dfs, axis=1)
```

```python
df
```

```python
df['Date'] = dfY['Date']
df['y'] = dfY['y']
```

```python
features = list(df.columns.values)[0:-2] # remove the last two columns which are not the main features ('Date' and 'y')
```

```python
df
```

```python
train, valid, test = TSDataset.from_pandas(df, dt_col='Date', target_col='y', extra_feature_col=features, with_split=True, val_ratio=0.1, test_ratio=0.1)
```

```python
valid.df
```

```python
lookback, horizon = 7, 1
```

```python
scaler = StandardScaler()
for data in [train, valid, test]:
    data.deduplicate() \
        .impute() \
        .gen_dt_feature() \
        .scale(scaler, fit=(data is train)) \
        .roll(lookback=lookback, horizon=horizon)
```

```python
X, y = train.to_numpy()
X.shape, y.shape
```

```python
from bigdl.chronos.forecaster.lstm_forecaster import LSTMForecaster
```

```python
from torch import nn
```

```python
import torch
```

```python
extreme_weight = torch.full([1], 2.0)
```

```python
loss = nn.BCEWithLogitsLoss(pos_weight=extreme_weight)
```

```python
lstm = LSTMForecaster(past_seq_len=lookback, input_feature_num=X.shape[-1], output_feature_num=y.shape[-1], hidden_dim=8, layer_num=2, \
                      optimizer='Adam', loss=loss, lr=1e-5)
```

```python
lstm.fit((X, y), epochs=10)
```

```python
X_valid, y_valid = valid.to_numpy()
```

```python
y_valid = valid.unscale_numpy(y_valid)
```

```python
lstm.evaluate((X_valid, y_valid))
```

```python
prediction_valid = lstm.predict(X_valid)
```

```python
p_valid = valid.unscale_numpy(prediction_valid)
```

```python
plt.figure(figsize=(24,6))
plt.plot(p_valid[:,:,0][0:364])
plt.plot(y_valid[:,:,0][0:364])
plt.legend(["prediction (LSTM)", "ground truth"])
plt.title('Extreme Precipitation on the Test Dataset (Muong Te)')
plt.xlabel('Day')
plt.ylabel('Extreme')
```

```python
def classify(p, epsilon=0.5):
    if (p >= epsilon):
        return 1.0
    else:
        return 0.0
```

```python
z_valid = [classify(p) for p in p_valid.flatten()]
```

```python
import sklearn
import sklearn.metrics
```

```python
# AUC score
fpr, tpr, thresholds = sklearn.metrics.roc_curve(y_valid.flatten(), z_valid)
sklearn.metrics.auc(fpr, tpr)
```

```python
print(sklearn.metrics.classification_report(y_valid.flatten(), z_valid))
```
